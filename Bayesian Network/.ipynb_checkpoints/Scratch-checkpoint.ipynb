```python
from importlib import reload  
import my_network
reload(my_network)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
X_train = np.loadtxt("Data/AMZN_20012020/X_train.txt")
y_train = np.loadtxt("Data/AMZN_20012020/y_train.txt")
X_test = np.loadtxt("Data/AMZN_20012020/X_test.txt")
y_test = np.loadtxt("Data/AMZN_20012020/y_test.txt")

train_data = (X_train, y_train)
test_data = (X_test, y_test)

input_scaler = StandardScaler()
input_scaler.fit(train_data[0])
X_train = input_scaler.transform(train_data[0])
X_test = input_scaler.transform(test_data[0])

output_scaler = StandardScaler()
output_scaler.fit(train_data[1])
y_train = output_scaler.transform(train_data[1])
y_test = output_scaler.transform(test_data[1])
```

```python
bayes_network = my_network.BayesianNetwork([X_train.shape[1], 10, 10, y_train.shape[1]], 
                                           ["sigmoid", "sigmoid", "linear"], nu1=0.01, nu2=0.01,
                                           learning_rate=0.1, use_langevin=True)

output = bayes_network.mcmc_sampler(X_train, y_train, X_test, y_test, update=True, n_samples=1000)
w_samples, b_samples, tau2_samples, train_loss, test_loss, accepted = output
```

```python
plt.plot(train_loss, label="train")
plt.plot(test_loss, label="test")
plt.legend()
```

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
i = Input(shape=X_train[1].shape)
x = Dense(10, activation="sigmoid")(i)
x = Dense(10, activation="sigmoid")(x)
o = Dense(y_train.shape[1], activation="linear")(x)
model = Model(i, o)
model.compile(optimizer="adam", loss='mse')
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100)
```
